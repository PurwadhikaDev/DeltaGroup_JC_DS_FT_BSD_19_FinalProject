# BUSINESS PROBLEM STATEMENT

## Context

Sebuah bank di portugal telah melakukan kampanye pemasaran(telemarketing) kepada para nasabahnya untuk produk deposito berjangka. Deposito berjangka adalah jenis akun tabungan yang membayar bunga tetap dengan jangka waktu tertentu. Bank sering menggunakan kampanye telemarketing untuk mempromosikan deposito berjangka kepada pelanggan mereka. Tujuan dari kampanye ini adalah untuk meyakinkan pelanggan untuk berlangganan deposito berjangka, yang dapat menguntungkan baik bagi bank maupun pelanggan. Deposito berjangka memberikan sumber pendanaan yang stabil bagi bank yang dapat digunakan untuk memberikan pinjaman dan berinvestasi di bidang lain dari bisnis mereka. Hal ini dapat membantu bank menjaga stabilitas keuangan dan memberikan layanan yang lebih baik kepada pelanggan mereka. Ketika pelanggan berlangganan deposito berjangka, hal ini dapat memberikan kesempatan bagi bank untuk menjual produk dan layanan keuangan lainnya kepada pelanggan, seperti kartu kredit, hipotek, atau rekening investasi.

Agar produk deposito dapat dipasarkan dengan efektif, bank perlu mempertimbangkan beberapa strategi pemasaran seperti iklan di media sosial, email pemasaran, atau pemasaran mesin pencari. Bank juga dapat mempertimbangkan untuk mengadakan acara atau sponsor acara untuk membangun kesadaran merek dan memperkenalkan produk deposito mereka. Selain itu, bank juga dapat menggunakan telemarketing untuk menghubungi pelanggan potensial secara langsung dan menjelaskan manfaat dari produk deposito mereka. Dalam memasarkan produk deposito, penting bagi bank untuk mempertimbangkan kebutuhan dan preferensi pelanggan serta mencari cara untuk membedakan diri dari pesaing.

## Problem Statement

### Data Analytics
Ada beberapa hal yang perlu dicari tahu untuk mengidentifikasi karakteristik dari nasabah.
1. Bagaimana faktor-faktor demografis dan ekonomi mempengaruhi nasabah untuk berlangganan deposito berjangka?
2. Saluran komunikasi apa yang paling efektif untuk menawarkan nasabah membuka rekening deposito?
3. Bagaimana lama durasi mempengaruhi keputusan nasabah untuk berlangganan deposito berjangka?
4. Bagaimana atribut socio ekonomi mempengaruhi keputusan nasabah untuk berlangganan deposito berjangka?

### Machine Learning
Dapatkah membangun machine learning model untuk memprediksi apakah seorang klien akan berlangganan deposito berjangka berdasarkan data demografi, keuangan, dan perilaku mereka, dan dapatkah model ini digunakan untuk meningkatkan kampanye pemasaran bank?

Untuk menyelesaikan masalah ini, kami akan menggunakan teknik supervised algoritma seperti algoritma klasifikasi untuk melatih model pada data historis dan memprediksi kemungkinan seorang klien akan berlangganan deposito berjangka, agar kemudian dapat menggunakan model ini untuk mengidentifikasi klien yang lebih mungkin berlangganan deposito berjangka dan menyesuaikan kampanye pemasaran bank untuk menargetkan klien tersebut.

## Goals

Berdasarkan dengan problem yang dirasakan oleh pihak Bank, maka pihak bank ingin melakukan analisis untuk mengidentifikasi karakteristik dari nansabah berdasarkan faktor demografis, saluran komunikasi yang efektif, durasi lama telfon, serta atribut socio ekonomi. Tujuannya jelas untuk memetakan nasabah yang sudah ada dan dapat membuat sebuah rekomendasi untuk nasabah nasabah baru nantinya. 

Selain itu, karena adanya kekhawatiran tentang nasabah baru tersebut, pihak Bank hendak membangun sebuah model machine learning yang dapat secara akurat memprediksi apakah pelanggan akan melakukan deposito berjangka atau tidak sehingga segala usaha kampanye pemasaran dapat difokuskan kepada para calon pelanggan yang lebih cenderung untuk berlangganan deposito berjangka. Pihak bank juga hendak mengidentifikasi beberapa faktor yang berkemungkinan menjadi kunci dalam mempengaruhi keputusan nasabah untuk melakukan deposito berjangka pada bank.

## Analytic Approach

Pendekatan yang dilakukan akan melibatkan analisis data eksplorasi untuk mendapatkan insights tentang hubungan antara variabel yang berbeda dengan variabel target. Kemudian data akan diproses dengan menangani missing value, encoding beberapa variabel, dan scaling variabel numerik tertentu. Kemudian akan dibangun model prediksi klasifikasi, seperti LogisticRegression dan DecisionTree, untuk memprediksi kemungkinan nasabah membuka deposito berjangka. Kemudian performa model akan dievaluasi menggunakan metric evaluation. Terakhir, akan diidentifikasi faktor utama yang memengaruhi keputusan nasabah untuk melakukan deposito berjangka bank dengan menganalisis feature importance model yang dibuat.

## Metric Evaluation

```python
# from IPython.display import Image
# Image(filename='./89BXgc.png')
```

Type 1 error: False Positives (Nasabah yang diprediksi berlangganan deposito berjangka namun aslinya tidak berlangganan)

Consequences: sumber daya yang terbuang, seperti waktu dan uang yang dihabiskan untuk kampanye pemasaran

Type 2 errors: False Negatives (Nasabah yang diprediksi tidak berlangganan deposito berjangka namun aslinya berlangganan)

Consequence: kehilangan sumber pendanaan yang stabil dan kehilangan peluang untuk penjualan silang produk keuangan lainnya

Untuk meminimalkan biaya kampanye pemasaran dengan tetap mempertahankan jumlah potensi pelanggan baru yang cukup bagi bank, tujuan kami adalah mengembangkan model yang dapat efektif mengurangi biaya pemasaran. Untuk mencapai tujuan ini, diperlukan keseimbangan antara presisi dan recall kelas positif. Maka dari itu, metrik evaluasi yang digunakan yaitu roc_auc.

### Data Correlation

#### Nominal scale correlation

```python
# Memeriksa korelasi menggunakan metode cramers
cat = df.select_dtypes('object').columns.to_list()

assoc_cramers = []
for col in  cat:
    assoc = round(cramers_v(df['y'], df[col]), 2)
    assoc_cramers.append(assoc)

df_cramers_v = pd.DataFrame(data = [assoc_cramers], columns = cat, index=['y'])

plt.figure(figsize = (13, 1))
sns.heatmap(df_cramers_v, annot = True, cmap = sns.color_palette('rocket_r'));
```

#### Ratio scale correlation

```python
num = df.select_dtypes(['int64', 'float64']).columns.to_list()

assoc_corratio = []
for col in  num:
    assoc = round(correlation_ratio(df['y'], df[col]), 2)
    assoc_corratio.append(assoc)

df_corr_ratio = pd.DataFrame(data = [assoc_corratio], columns = num, index = ['y'])

plt.figure(figsize = (13, 1))
sns.heatmap(df_corr_ratio, annot = True, cmap = sns.color_palette('rocket_r'));
```

Selanjutnya, berikut adalah nilai korelasi antara target y dengan fitur numerik. Terlihat bahwa terdapat korelasi yang sedang, yaitu pada fitur duration, pdays, euribor3m, dan nr.employed. Sisanya fitur yang lain memiliki korelasi yang lemah.

## Data Cleaning

### Missing value

```python
df.isna().sum()
```

```python
cat_col = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome']

for col in cat_col:
    print(f'{col}: {df[col].unique()}')
```

Seperti yang terlihat pada tabel, meskipun tidak terpadat NaN value pada semua fitur yang ada, namun apabila diperhatikan setiap unique value pada fitur kategorik, terdapat beberapa yang tidak diketahui nilainya, yaitu ditulis sebagai 'unknown'. Dalam hal ini, nilai tersebut akan dianggap sebagai missing value dan akan dilakukan treatment terhadapnya. Pertama, setiap nilai yang berisi nilai 'unknown' diganti dengan NaN, agar terbaca sebagai missing value.

```python
for col in cat_col:
    df[col] = df[col].replace('unknown', np.NaN)

df.isna().sum()
```

Terlihat bahwa pada fitur-fitur 'job', 'marital', 'education', 'default', 'housing', dan 'loan' menjadi memiliki missing value. Selanjutnya, dilihat kembali presentase dari missing value tersebut dibandingkan seluruh data.

```python
pd.set_option('display.max_colwidth', None)
dataDesc = []

for col in df.columns:
    dataDesc.append([
        col, 
        df[col].dtypes,
        df[col].isna().sum(),
        round((((df[col].isna().sum())/len(df))*100), 2),
        df[col].nunique(),
        df[col].sample(20).drop_duplicates().values
    ])

pd.DataFrame(dataDesc, columns = [
    "Data Features", 
    "Data Types",
    "Null",
    "Null Percentage",
    "Unique",
    "Unique Sample"
])
```

Dari fitur yang memiliki missing value, terlihat bahwa presentasenya hanyalah kecil, tidak lebih dari 5%, kecuali pada fitur 'default' yang memiliki missing value sebesar 20.87%.

### Anomaly

```python
df['default'].value_counts(dropna = False)
```

Apabila diperhatikan proporsi dari fitur 'default', terlihat bahwa nilai NaN tidaklah sedikit. Oleh karena itu, agar mengurangi bias pada data, kolom ini nanti akan dianalisa lebih lanjut di EDA.

## Model Benchmarking: SKFold

```python
# Algoritma yang akan digunakan
lr = LogisticRegression()
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier(random_state=2023)
rf = RandomForestClassifier(random_state=2023)
xgb = XGBClassifier(random_state=2023)
lgbm = lgb.LGBMClassifier()

# Kandidat algoritma
models = [lr,knn,dt,rf,xgb,lgbm]
score=[]
rata=[]
std=[]

for i in models:
    skfold=StratifiedKFold()
    estimator=Pipeline([
        ('preprocess',transform),
        ('sampler', SMOTE(sampling_strategy='minority', random_state=2023)),
        ('model',i)])
    model_cv=cross_val_score(estimator,x_train,y_train,cv=skfold,scoring='roc_auc')
    score.append(model_cv)
    rata.append(model_cv.mean())
    std.append(model_cv.std())
    
pd.DataFrame({'model':['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest', 'XGBoost', 'LightGBM'],'mean roc_auc':rata,'sdev':std}).set_index('model').sort_values(by='mean roc_auc',ascending=False)
```

Pada benchmarking menggunakan SKFold, ditemukan bahwa jenis model LightGBM yang memiliki nilai terbaik

## Model Benchmarking: Test Data

```python
score_roc_auc = []

def y_pred_func(i):
    estimator=Pipeline([
        ('preprocess',transform),
        ('sampler', SMOTE(sampling_strategy='minority',random_state=2023)),
        ('model',i)])
    x_train,x_test
    
    estimator.fit(x_train,y_train)
    return(estimator,estimator.predict(x_test),x_test)

for i,j in zip(models, ['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest', 'XGBoost','LightGBM']):
    estimator,y_pred,x_test = y_pred_func(i)
    y_predict_proba = estimator.predict_proba(x_test)[:,1]
    score_roc_auc.append(roc_auc_score(y_test,y_predict_proba))
    print(j,'\n', classification_report(y_test,y_pred))
    
pd.DataFrame({'model':['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest', 'XGBoost','LightGBM'],
             'roc_auc score':score_roc_auc}).set_index('model').sort_values(by='roc_auc score',ascending=False)
```

Pada benchmarking model menggunakan test data juga ditemukan bahwa jenis model LightGBM memiliki nilai terbaik

## Hyperparameter Tuning

Hyperparameter tuning untuk jenis model LightGBM dapat melihat referensi sebagai berikut:
- https://lightgbm.readthedocs.io/en/latest/Parameters.html
- https://lightgbm.readthedocs.io/en/latest/Parameters-Tuning.html
- https://stackoverflow.com/questions/58815016/cross-validating-with-imblearn-pipeline-and-gridsearchcv
- https://scikit-learn.org/stable/auto_examples/compose/plot_compare_reduction.html#sphx-glr-auto-examples-compose-plot-compare-reduction-py

```python
# Define model dan pipeline
lgbm = lgb.LGBMClassifier()
sm=SMOTE(sampling_strategy='minority', random_state=2023)

estimator=Pipeline([
    ('preprocess',transform),
    ('sampler',sm),
    ('model',lgbm)
])

# Define parameter yang digunakan
hyperparam_space=[{
    'model__max_bin': [245, 250, 255, 260, 265],
    'model__num_leaves':[27, 29, 31, 33, 36],
    'model__min_data_in_leaf': [30, 35, 40, 45],
    'model__num_iterations':[65,70,75,80,85],
    'model__learning_rate': [0.1, 0.075, 0.125, 0.05],
    'model__random_state': [2023]
}]
```

```python
# Tuning menggunakan GridSearchCV
grid = GridSearchCV(estimator, n_jobs=-1, param_grid=hyperparam_space, scoring='roc_auc', cv=StratifiedKFold(shuffle=True, random_state=2023))

# Fitting data training untuk mencari parameter terbaik
grid.fit(x_train,y_train)
print(grid.best_score_)
print(grid.best_params_)
```

Hyperparameter terbaik dari gridsearchcv yang dilakukan yaitu `learning_rate` = 0.05, `max_bin` = 265, `min_data_in_leaf` = 30, `num_iterations` = 65, dan `num_leaves` = 31

Parameter yang ditentukan oleh hyperparam_space masih bisa diubah lagi untuk menemukan model yang paling bagus, namun karena keterbatasan kemampuan komputasi, hanya digunakan yang telah didefinisikan diatas.

```python
# Fitting data training menggunakan parameter terbaik yang berhasil ditentukan di proses sebelumnya
best_model = grid.best_estimator_
best_model.fit(x_train, y_train)
```

```python
# Fitting data training menggunakan estimator
estimator=Pipeline([
    ('preprocess',transform),
    ('sampler',sm),
    ('model',lgbm)
])
estimator.fit(x_train, y_train)
```
